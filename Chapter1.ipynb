#### 1.5 넘파이

```python
# 넘파이 산술 연산 예시
import numpy as np

x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print("x + y:\n", x + y)
print("x - y:\n", x - y)
print("x * y:\n", x * y)
print("x / y:\n", x / y)
```

```python
# 브로드캐스트: 넘파이 배열과 스칼라값의 조합으로 된 산술 연산
x = np.array([1.0, 2.0, 3.0])
print(x / 2.0)
```

```python
# 브로드캐스트: 형상이 다른 배열끼리의 연산
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
A * B
```

#### 1.6 matplotlib

```python
# matplotlib의 pyplot 모듈로 sin 함수 그리기
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# cos 함수도 추가로 그리기
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```

```python
# 이미지 표시하기
from matplotlib.image import imread

img = imread("cactus.jpg")
plt.imshow(img)
plt.show()
```
